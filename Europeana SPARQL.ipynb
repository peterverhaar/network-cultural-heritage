# Europeana SPARQL

The query below shows a number of items contributed by the Rijksmuseum. Can you request items from the following institutions?

* "The National Archives of Norway"
* "The British Museum and The Portable Antiquities Scheme"
* "Nationaal Archief, Den Haag"
* "UCL Museums"


```python
from SPARQLWrapper import SPARQLWrapper, JSON

endpoint = 'http://sparql.europeana.eu/'
```

```python
query = """
PREFIX dc: <http://purl.org/dc/elements/1.1/>
PREFIX edm: <http://www.europeana.eu/schemas/edm/>
PREFIX ore: <http://www.openarchives.org/ore/terms/>
PREFIX html: <http://www.w3.org/1999/xhtml/vocab#>

SELECT ?title ?img_url ?creator
WHERE {
?object ore:proxyIn ?local_aggr .
?object dc:title ?title .
OPTIONAL{ ?object dc:creator ?creator . } .
?local_aggr edm:object ?img_url .
?local_aggr edm:dataProvider "Rijksmuseum" .
}
LIMIT 40
"""


sparql = SPARQLWrapper(endpoint)
sparql.setQuery(query)
sparql.setReturnFormat(JSON)
data = sparql.query().convert()
```

```python
import pandas as pd
import numpy as np 

df = pd.json_normalize(data['results']['bindings'])  


for i,row in df.iterrows():
    print( f"{row['title.value']}" )
    print( f"{row['img_url.value']}\n" )
```

## Selection based on the country of GLAM institutions

The query below displays items managed by GLAM institutions located in Italy. 

Can you change the query in such a way that it displays items that can be found in one of the following countries?

* "france"
* "germany"
* "united kingdom"
* "netherlands"
* "czech republic"
* "belgium"
* "greece"
* "estonia"
* "spain"


```python
query = """
PREFIX dc: <http://purl.org/dc/elements/1.1/>
PREFIX edm: <http://www.europeana.eu/schemas/edm/>
PREFIX ore: <http://www.openarchives.org/ore/terms/>
PREFIX html: <http://www.w3.org/1999/xhtml/vocab#>

SELECT ?title ?creator ?glam_inst ?img_url
WHERE {
?object edm:type "IMAGE" .
?object dc:title ?title .
?object dc:creator ?creator .
?object ore:proxyIn ?local_aggr .
OPTIONAL{ ?local_aggr edm:object ?img_url . } .
?local_aggr edm:provider ?glam_inst .

?object ore:proxyFor ?cho .
?eur_aggr edm:aggregatedCHO ?cho .
?eur_aggr a edm:EuropeanaAggregation .

?eur_aggr edm:country 'italy' .
}
LIMIT 100

"""


sparql = SPARQLWrapper(endpoint)
sparql.setQuery(query)
sparql.setReturnFormat(JSON)
data = sparql.query().convert()
```

```python
import pandas as pd
import numpy as np 
from IPython.core.display import display, HTML


df = pd.json_normalize(data['results']['bindings'])  

for i,row in df.iterrows():
    print( f"{row['title.value']}" )
    print( f"{row['creator.value']}" )
    img = row['img_url.value']
    display(HTML(f'<a target="_new" href="{img}"><img src="{img}" style="width: 200px;"/></a>'))

```

## Licences

Which licences are most commonly used for images? 

```python
query = """
PREFIX dc: <http://purl.org/dc/elements/1.1/>
PREFIX edm: <http://www.europeana.eu/schemas/edm/>
PREFIX ore: <http://www.openarchives.org/ore/terms/>
PREFIX html: <http://www.w3.org/1999/xhtml/vocab#>

SELECT ?rights (COUNT(?rights) as ?n)
WHERE {
?object edm:type "IMAGE" .
?object ore:proxyIn ?local_aggr .
?local_aggr edm:rights ?rights .
}
GROUP BY ?rights
ORDER BY DESC(?n)
"""


sparql = SPARQLWrapper(endpoint)
sparql.setQuery(query)
sparql.setReturnFormat(JSON)
data = sparql.query().convert()


df = pd.json_normalize(data['results']['bindings'])  

for i,row in df.iterrows():
    print( f"{row['rights.value']} : {row['n.value']}" )

```
